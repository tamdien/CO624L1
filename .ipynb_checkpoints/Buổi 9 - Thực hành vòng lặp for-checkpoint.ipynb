### B√†i 6

C·∫ßn c√≥ t·ªïng 200.000ƒë t·ª´ 3 lo·∫°i gi·∫•y b·∫°c 1000ƒë, 2000ƒë, v√† 5000ƒë. L·∫≠p 
ch∆∞∆°ng t√¨nh ƒë·ªÉ t√¨m t·∫•t c·∫£ c√°c ph∆∞∆°ng √°n c√≥ th·ªÉ.

```python
tong_tien = 200000
count = 0
for tien_1000ƒë in range(0, tong_tien + 1, 1000):
    for tien_2000ƒë in range(0, tong_tien + 1, 2000):
        for tien_5000ƒë in range(0, tong_tien + 1, 5000):
            tong = tien_1000ƒë + tien_2000ƒë + tien_5000ƒë
            if tong == tong_tien:
                print("Ph∆∞∆°ng √°n:", tien_1000ƒë // 1000, "t·ªù 1000ƒë",tien_2000ƒë // 2000, "t·ªù 2000ƒë", tien_5000ƒë // 5000, "t·ªù 5000ƒë")
                count += 1

print(count)
```

```python
count = 0
for i in range(200+1):
    for j in range (100+1):
        for k in range (40+1):
            if (i*1000 + j*2000 + k*5000) == 200000:
                print("S·ªë t·ªù gi·∫•y b·∫°c 1000:",i,",","S·ªë t·ªù gi·∫•y b·∫°c 2000:",j,"," "S·ªë t·ªù gi·∫•y b·∫°c 5000:",k)
                count += 1

print(count)
```
